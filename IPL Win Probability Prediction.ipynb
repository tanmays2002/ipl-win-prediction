# Run Left To Chase

```python
# To fond total scores in first innings
```

```python
total_score_df = df.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df = total_score_df[total_score_df['inning'] == 1]
total_score_df.head(5)
```

```python
# Merge this columns with match dataset
match_df = match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
match_df.sample(5)
```

```python
# There are some teams which are not playing in current IPL
match_df['team1'].unique()
```

```python
#We have to replace the team names with current one
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
# We are filtering this teams which are playing current season of IPL
teams = ['Sunrisers Hyderabad',
         'Mumbai Indians',
         'Royal Challengers Bangalore',
         'Kolkata Knight Riders',
         'Kings XI Punjab',
         'Chennai Super Kings',
         'Rajasthan Royals',
         'Delhi Capitals'
        ]


match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

#
In this deliveries dataset, there is a column having values
of duckwarth luies method(scores updated after rain)
we have to remove this values

```python
match_df['dl_applied'].value_counts()
```

```python
# We are extracting those macthes where dl method is not applied.
match_df = match_df[match_df['dl_applied']==0]
match_df.head()
```

```python
# Let's merge this dataset with deliveries on match_id
delivery_df = match_df.merge(df,on='match_id')
```

```python
# We want only thos rows having second inning info
delivery_df = delivery_df[delivery_df['inning']==2]
delivery_df.head(5)
```

### Runs Left

Runs Left = total runs - current score

```python
# To find Current Score
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
#Runs Left = total runs - current score
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

### Balls Left 

Balls Left = 126 - (over*6 + ball)

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
delivery_df.head(5)
```

```python
delivery_df['player_dismissed'].sample(5)
```

#
Problem with this column is there is nan values and player name whose wicket is down so we have to convert this column into '0' and '1' for the count of wickets

```python
#Filling nan values with 0
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna('0')
```

```python
# Wicket =1 otherwise o
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x=='0'else'1')
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
```

```python
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
```

```python
delivery_df['wickets'] = 10 - wickets
delivery_df.sample(5)
```

### Current Run Rate
       = current runs / current over

```python
delivery_df['crr'] =  (delivery_df['current_score']*6) / (120 - delivery_df['balls_left'])    
```

# 

# Required Run Rate
          = required runs / remaing balls

```python
delivery_df['rrr'] = (delivery_df['runs_left']*6) / delivery_df['balls_left']
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0

delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
# Shuffling the Rows
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.describe()
```

```python
# There are some inf values present in row cause of last ball of inning
final_df = final_df[final_df['balls_left'] !=0]
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
```

```python
# train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_test
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('OHE',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
